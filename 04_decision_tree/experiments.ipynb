```python
iris = '../data/iris/bezdekIris.data'
iris = pd.read_csv(iris)
iris.columns = ['petal length', 'petal width', 'sepal length', 'sepal width', 'class']
```

---

```python
from decision_tree import DecisionTree

X = iris[['petal length', 'petal width', 'sepal length', 'sepal width']]
y = iris['class']

for i in range(1, 8):
    test = DecisionTree()
    test.fit(X, y, max_depth=i)
    y_pred = test.predict(X)
    correct = sum(1 for p, r in zip(y_pred, y) if p == r)
    print(f'Max depth: {i} Accuracy: {correct / len(y)}')
```
