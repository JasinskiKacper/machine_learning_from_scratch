```python
X = student[['parental level of education', 'race/ethnicity']]
X.head()
```

```python
y = student['math score'] > 65
y.head()
```

```python
from decision_tree import gini
print(gini(y))
```

```python
unique = X['parental level of education'].unique()
unique

X['parental level of education'].iloc[0]
```

```python
def split_left_right(feature: pd.Series, 
                     y: pd.Series, 
                     condition: str|int) -> tuple[list, list]:
    if len(feature) != len(y):
        return None
    
    n = len(y)
    left = []
    right = []
    for i in range(n):
        if type(condition) == str:
            if feature[i] == condition:
                left.append(y[i])
            else:
                right.append(y[i])

        else:
            if feature[i] <= condition:
                left.append(y[i])
            else:
                right.append(y[i])

    return (pd.Series(left), pd.Series(right), n)

left, right, n = split_left_right(X['parental level of education'], y, condition='some college')

print(gini(left), gini(right), n)
    
```

```python
min_gini = []
for i in range(len(unique)):
    uni = unique[i]

    left, right, n = split_left_right(X['parental level of education'],
                                      y,
                                      uni)
    weighted_gini = (len(left) / n * gini(left)) + (len(right) / n * gini(right))
    min_gini.append(weighted_gini)

min_gini
```
