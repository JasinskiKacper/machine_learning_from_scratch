Importing libraries

```python
import numpy as np
import pandas as pd
```

Importing data

```python
import os

data = os.path.abspath('../data/spam.csv')
df = pd.read_csv(data, encoding='latin-1')

df = pd.DataFrame({'sms': df['v2'],'result': df['v1']})
```

---

Chcecking data

```python
df.head()
```

---

### **Multinomial NB**

```python
from multinomial_nb import Multinomial_nb

X = df['sms']
y = df['result']

multi_nb = Multinomial_nb()

multi_nb.fit(X, y)

X_predict = df['sms']
y_pred = multi_nb.predict(X_predict)

print(y_pred)
```

---

### **Gaussian NB**

```python
data = os.path.abspath('../data/iris/iris.data')
iris = pd.read_csv(data, header=None)

iris.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class']

iris
```

```python
from gaussian_nb import Gaussian_nb

X = iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
y = iris['class']

gauss_nb = Gaussian_nb()
gauss_nb.fit(X, y)

y_pred = gauss_nb.predict(X)
y_pred
```

```python
classes = gauss_nb.predict_class(X)

classes
```
