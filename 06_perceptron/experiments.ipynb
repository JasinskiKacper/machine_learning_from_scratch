### **Rosenblatt's perceptron**

```python
from perceptron import Perceptron

X = df[['sepal length', 'sepal width', 'petal length', 'petal width']].to_numpy()
y = df[['class']]

y = y['class'].map({'Iris-setosa': 1,
           'Iris-versicolor': -1,
           'Iris-virginica': -1
           })
y = y.to_numpy()

net = Perceptron()

net.fit(X, y, lr=0.05, max_iters=100, stats=True)

y_pred = net.predict(X)
```
