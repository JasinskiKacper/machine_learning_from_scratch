```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

Importing data

```python
path = os.path.abspath('../data/iris/iris.data')
df = pd.read_csv(path, header=None)
```

Checking data

```python
df.columns = ['sepal length', 'sepal width', 'petal length', 'petal width', 'class']

df = df[['sepal length', 'sepal width', 'petal length', 'petal width']]

df
```

---

Visualizing data

```python
sns.pairplot(df)
```

---

### **Kmeans (Lloyd's algorithm)**

```python
from standard_kmeans import Standard_Kmeans
X = df[['sepal length', 'sepal width']].to_numpy()
std_kmeans = Standard_Kmeans()

k=3
std_kmeans.fit(X, k=k, max_iter=500)

centroids = std_kmeans.centroid
print(centroids)

colors = ['tab:green', 'tab:red', 'tab:blue', 'tab:gray', 'tab:orange', 'tab:purple', 'tab:brown']
fig, ax = plt.subplots()
for i in range(k):
    cluster = X[std_kmeans.labels == i]

    ax.scatter(x=cluster[:, 0], y=cluster[:, 1], c=colors[i])
    ax.scatter(x=centroids[i][0], y=centroids[i][1], c=colors[i])

ax.grid()
ax.set_title('K-means using Lloyd s algorithm')
ax.set_ylabel('sepal width')
ax.set_xlabel('sepal length')

print(std_kmeans.predict(np.array([5.0, 3.5])))
```

---

### **Kmeans++**

```python
from kmeans_pp import Kmeans_pp

X = df[['sepal length', 'sepal width']].to_numpy()

pp_kmeans = Kmeans_pp()

k=3
pp_kmeans.fit(X, k=k, max_iter=500)

centroids = pp_kmeans.centroid
print(centroids)

colors = ['tab:green', 'tab:red', 'tab:blue', 'tab:gray', 'tab:orange', 'tab:purple', 'tab:brown']
fig, ax = plt.subplots()
for i in range(k):
    cluster = X[pp_kmeans.labels == i]

    ax.scatter(x=cluster[:, 0], y=cluster[:, 1], c=colors[i])
    ax.scatter(x=centroids[i][0], y=centroids[i][1], c=colors[i])

ax.grid()
ax.set_title('K-means using Lloyd s algorithm')
ax.set_ylabel('sepal width')
ax.set_xlabel('sepal length')

print(pp_kmeans.predict(np.array([5.0, 3.5])))
```
